```python
import torch
from ultralytics import YOLO
from ultralytics import utils
from torchvision import models
from PIL import Image
import os
```

```python
dataset = 'Autogathered_Dataset/2 (labeless removed)/dataset.yaml'
output_path = 'Autogathered_Dataset/2 (labeless removed)/'
output_dir = 'runs/'
```

```python
print(torch.cuda.device_count())
```

```python
# Load a pretrained YOLO model (recommended for training)
model = YOLO('yolov8n.pt')
```

```python
# Train the model using the 'coco128.yaml' dataset for 3 epochs
results = model.train(data=dataset, epochs=10, device=0, batch=8, project=output_path, name=output_dir)

# Evaluate the model's performance on the validation set
results = model.val()
```

```python
model = YOLO(output_path + 'runs/weights/best.pt')

# Perform object detection on an image using the model
#results = model.predict(output_path + 'Images/0.png')

screw_image = 'C:/Users/drago/Documents/gitrepos/Cross-Recessed-Screw_Deep-Learning-Datasets/Training_Images/000016.jpg'
results_screw_data = model.predict(screw_image)
```

```python
# Show the results
for r in results_screw_data:
    im_array = r.plot()  # plot a BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
    im.show()  # show image
    im.save('results.jpg')  # save image
```

# Label images within the small_screw_training data

```python
small_screw_dataset = 'C:/Users/drago/Documents/gitrepos/Cross-Recessed-Screw_Deep-Learning-Datasets/Small_Training_Images/Images/'
output_dir = 'C:/Users/drago/Documents/gitrepos/Cross-Recessed-Screw_Deep-Learning-Datasets/Small_Training_Images/Predictions-example_labels/'


model = YOLO(output_path + 'runs/weights/best.pt')

# Iterate through all images in the folder and output predictions as example pre-labels
for file in os.listdir(small_screw_dataset):
    # Run model on images and output labelled images to output folder
    prediction = model.predict(small_screw_dataset + file)
    # Save the results
    for r in prediction:
        im_array = r.plot()  # plot a BGR numpy array of predictions
        im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
        im.save(output_dir + file)  # save image
```
