```python
small_screw_dataset = 'C:/Users/drago/Documents/gitrepos/Cross-Recessed-Screw_Deep-Learning-Datasets/Small_Training_Images/Images/'
output_dir = 'C:/Users/drago/Documents/gitrepos/Cross-Recessed-Screw_Deep-Learning-Datasets/Small_Training_Images/Predictions-example_labels/'


model = YOLO(output_path + 'runs/weights/best.pt')

# Iterate through all images in the folder and output predictions as example pre-labels
for file in os.listdir(small_screw_dataset):
    # Run model on images and output labelled images to output folder
    prediction = model.predict(small_screw_dataset + file)
    # Save the results
    for r in prediction:
        im_array = r.plot()  # plot a BGR numpy array of predictions
        im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
        im.save(output_dir + file)  # save image
```
